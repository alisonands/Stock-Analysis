```python
import matplotlib.pyplot as plt
import seaborn as sbn
import pandas as pd
import scipy.stats as st
```

```python
data_df = pd.read_csv("data_sets/tech_data.csv")
data_df['Industry'].unique()
```

```python
data_df['Date'] = pd.to_datetime(data_df['Date'])
```

```python
df1 = data_df.loc[data_df['Industry']=="Semiconductors",:]
df2 = data_df.loc[data_df['Industry']=="Software",:]
df3 = data_df.loc[data_df['Industry']=="Consumer",:]
df4 = data_df.loc[data_df['Industry']=="Infotech",:]
```

```python
group_df1 = df1.groupby("Date")['Close'].mean().reset_index()
group_df2 = df2.groupby("Date")['Close'].mean().reset_index()
group_df3 = df3.groupby("Date")['Close'].mean().reset_index()
group_df4 = df4.groupby("Date")['Close'].mean().reset_index()
```

```python
merged_df = pd.merge(group_df1, group_df2, on='Date', how='inner', suffixes=('_df1', '_df2'))
merged_df = pd.merge(merged_df, group_df3, on='Date', how='inner', suffixes=('_merged', '_df3'))
merged_df = pd.merge(merged_df, group_df4, on='Date', how='inner', suffixes=('_df3', '_df4'))
```

```python
merged_df = merged_df.rename(columns={"Close_df1": "Semiconductors",
                                      "Close_df2": "Software",
                                      "Close_df3": "Consumer",
                                      "Close_df4": "Infotech"})
```

```python
merged_df = merged_df.dropna()
merged_df
```

```python
close_columns = ['Semiconductors', 'Software', 'Consumer', 'Infotech']
correlation_matrix = merged_df[close_columns].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(6,5))
sbn.heatmap(correlation_matrix, annot = True)
plt.xticks(rotation = 45, ha='right')
plt.title('Correlation of Closing Price')
```

```python
slope1, intercept1, rvalue1, pvalue1, stderr1 = st.linregress(merged_df['Semiconductors'], merged_df['Software'])

line_eq1 = f"y = {slope1:.2f}x + {intercept1:.2f}"

regressed_values1 = (slope1 * merged_df['Semiconductors']) + intercept1

x_axis = merged_df['Semiconductors']
y_axis = merged_df['Software']
plt.scatter(x_axis, y_axis)
plt.plot(x_axis, regressed_values1, color='r')
plt.annotate(line_eq1, (min(x_axis), max(y_axis)-10), fontsize=15, color='red')
plt.xlabel("Closing Price of Semiconductors")
plt.ylabel("Closing Price of Software")
correlation_coefficient1 = f'The r-value is: {rvalue1:.2f}'
print(correlation_coefficient1)
```

```python
slope2, intercept2, rvalue2, pvalue2, stderr2 = st.linregress(merged_df['Semiconductors'], merged_df['Infotech'])

line_eq2 = f"y = {slope2:.2f}x + {intercept2:.2f}"

regressed_values2 = (slope2 * merged_df['Semiconductors']) + intercept2

x_axis = merged_df['Semiconductors']
y_axis = merged_df['Infotech']
plt.scatter(x_axis, y_axis)
plt.plot(x_axis, regressed_values2, color='r')
plt.annotate(line_eq2, (min(x_axis), max(y_axis)-10), fontsize=15, color='red')
plt.xlabel("Closing Price of Semiconductors")
plt.ylabel("Closing Price of Infotech")
correlation_coefficient2 = f'The r-value is: {rvalue2:.2f}'
print(correlation_coefficient2)
```

```python
slope3, intercept3, rvalue3, pvalue3, stderr3 = st.linregress(merged_df['Software'], merged_df['Infotech'])

line_eq3 = f"y = {slope3:.2f}x + {intercept3:.2f}"

regressed_values3 = (slope3 * merged_df['Software']) + intercept3

x_axis = merged_df['Software']
y_axis = merged_df['Infotech']
plt.scatter(x_axis, y_axis)
plt.plot(x_axis, regressed_values3, color='r')
plt.annotate(line_eq3, (min(x_axis), max(y_axis)-10), fontsize=15, color='red')
plt.xlabel("Closing Price of Software")
plt.ylabel("Closing Price of Infotech")
correlation_coefficient3 = f'The r-value is: {rvalue3:.2f}'
print(correlation_coefficient3)
```
