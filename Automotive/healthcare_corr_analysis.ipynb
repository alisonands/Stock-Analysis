```python
import matplotlib.pyplot as plt
import seaborn as sbn
import pandas as pd
import scipy.stats as st
```

```python
data_df = pd.read_csv("data_sets/Healthcare_Stock_Data2.csv")
data_df['Industry'].unique()
```

```python
data_df['date'] = pd.to_datetime(data_df['date'], format='%m/%d/%Y')
data_df = data_df.rename(columns={'date':"Date",'1. open': "Open", '2. high': "High", '3. low': "Low", '4. close': "Close", '5. volume':"Volume"})
data_df
```

```python
df1 = data_df.loc[data_df['Industry']=="Medical Instruments & supplies",:]
df2 = data_df.loc[data_df['Industry']=="Healthcare plans",:]
df3 = data_df.loc[data_df['Industry']=="Biotechnology",:]
df4 = data_df.loc[data_df['Industry']=="Drug Manufacturers",:]
```

```python
group_df1 = df1.groupby("Date")['Close'].mean().reset_index()
group_df2 = df2.groupby("Date")['Close'].mean().reset_index()
group_df3 = df3.groupby("Date")['Close'].mean().reset_index()
group_df4 = df4.groupby("Date")['Close'].mean().reset_index()
```

```python
merged_df = pd.merge(group_df1, group_df2, on='Date', how='inner', suffixes=('_df1', '_df2'))
merged_df = pd.merge(merged_df, group_df3, on='Date', how='inner', suffixes=('_merged', '_df3'))
merged_df = pd.merge(merged_df, group_df4, on='Date', how='inner', suffixes=('_df3', '_df4'))
```

```python
merged_df = merged_df.rename(columns={"Close_df1": "Medical Instruments & supplies",
                                      "Close_df2": "Healthcare plans",
                                      "Close_df3": "Biotechnology",
                                      "Close_df4": "Drug Manufacturers"})
```

```python
merged_df = merged_df.dropna()
merged_df
```

```python
close_columns = ['Medical Instruments & supplies', 'Healthcare plans', 'Biotechnology', 'Drug Manufacturers']
correlation_matrix = merged_df[close_columns].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(6,5))
sbn.heatmap(correlation_matrix, annot = True)
plt.xticks(rotation = 45, ha='right')
plt.title('Correlation of Closing Price')
```

```python
slope1, intercept1, rvalue1, pvalue1, stderr1 = st.linregress(merged_df['Healthcare plans'], merged_df['Biotechnology'])

line_eq1 = f"y = {slope1:.2f}x + {intercept1:.2f}"

regressed_values1 = (slope1 * merged_df['Healthcare plans']) + intercept1

x_axis = merged_df['Healthcare plans']
y_axis = merged_df['Biotechnology']
plt.scatter(x_axis, y_axis)
plt.plot(x_axis, regressed_values1, color='r')
plt.annotate(line_eq1, (min(x_axis), max(y_axis)-10), fontsize=15, color='red')
plt.xlabel("Closing Price of Healthcare plans")
plt.ylabel("Closing Price of Biotechnology")
correlation_coefficient1 = f'The r-value is: {rvalue1:.2f}'
print(correlation_coefficient1)
```

```python
slope2, intercept2, rvalue2, pvalue2, stderr2 = st.linregress(merged_df['Healthcare plans'], merged_df['Drug Manufacturers'])

line_eq2 = f"y = {slope2:.2f}x + {intercept2:.2f}"

regressed_values2 = (slope2 * merged_df['Healthcare plans']) + intercept2

x_axis = merged_df['Healthcare plans']
y_axis = merged_df['Drug Manufacturers']
plt.scatter(x_axis, y_axis)
plt.plot(x_axis, regressed_values2, color='r')
plt.annotate(line_eq2, (min(x_axis), max(y_axis)-10), fontsize=15, color='red')
plt.xlabel("Closing Price of Healthcare plans")
plt.ylabel("Closing Price of Drug Manufacturers")
correlation_coefficient2 = f'The r-value is: {rvalue2:.2f}'
print(correlation_coefficient2)
```
