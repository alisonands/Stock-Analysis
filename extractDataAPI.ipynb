```python
#import dependencies
# !pip install yfinance
# !pip install pandas_datareader
import pandas as pd
import yfinance as yf
yf.pdr_override()
from pandas_datareader import data as wb
from datetime import date
```

```python
#create dataframe to hold the information of Tickers
info_df = pd.DataFrame({"Ticker" : ["WD", "PFSI", "LDI", "GHI", "COOP", "AFL", "HIG", "PRU", "ALL", "PGR", "MS", "BLK", "GS", "TROW", "BEN", "V", "MA", "AXP", "DFS", "COF"],
                        "Sector" : "",
                        "Industry" : "Finance"})
#populate the Sectors
info_df["Sector"].iloc[0:5] ="Mortgage"
info_df["Sector"].iloc[5:10] ="Insurance"
info_df["Sector"].iloc[10:15] ="Asset Management"
info_df["Sector"].iloc[15:20] ="Credit Services"

#Display dataframe
info_df
```

```python
#set end date for today's date for up to date data
end = date.today()
#List to append data temporarily
ls= []
#dataframe to hold all the retrieved data
final_df = {}

#loop through each ticker in info_df
for sym in info_df["Ticker"]:

    #store data retrieved a Ticker from 2018-01-01 to today's date
    stock_data = wb.DataReader({sym}, start='2018-1-1', end= end)
    stock_data["Ticker"] = sym
    
    #appennding the data in ls list
    ls.append(stock_data)
    
    #concate that ls to final_df dataframe
    final_df = pd.concat(ls)
#show the dataframe
final_df
```

```python
#merging the dataframes to show sector and industry
finance_df = pd.merge(final_df, info_df, on = "Ticker", how = "left")
finance_df = finance_df[["Ticker", "Sector", "Industry", "Open", "High", "Low", "Close", "Volume"]]
#show the result
finance_df
```

```python
#create a csvfile
finance_df.to_csv("Financial_stock_data.csv")
```
