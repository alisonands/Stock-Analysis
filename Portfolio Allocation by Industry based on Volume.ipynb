```python
from alpha_vantage.timeseries import TimeSeries
import matplotlib.pyplot as plt
import sys
import pandas as pd


# Read CSV
healthcare_data = pd.read_csv("Healthcare_Stock_Data4.csv")
healthcare_data
```

```python
# Assuming your DataFrame is named 'df'
# Replace 'df' with your actual DataFrame name if it's different

# Calculate the total volume for each industry
total_volume_by_industry = healthcare_data.groupby('Industry')['Volume'].sum()

# Calculate the percentage of total volume for each industry
volume_percentage_by_industry = (total_volume_by_industry / total_volume_by_industry.sum()) * 100

# Define colors for each industry
colors = plt.cm.Paired(range(len(volume_percentage_by_industry)))

# Explode a section for emphasis (optional)
explode = [0.1 if i == volume_percentage_by_industry.idxmax() else 0 for i in volume_percentage_by_industry.index]

# Plot a pie chart
plt.figure(figsize=(10, 6))
plt.pie(volume_percentage_by_industry, labels=volume_percentage_by_industry.index, autopct='%1.1f%%', startangle=140,
        colors=colors, explode=explode, shadow=True)
plt.title('Portfolio Allocation by Industry based on Volume')
plt.show()
```
