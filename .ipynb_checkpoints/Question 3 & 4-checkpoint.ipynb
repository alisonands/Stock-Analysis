```python
# Import dependencies
import matplotlib.pyplot as plt
import seaborn as sbn
import pandas as pd
import scipy.stats as st
import plotly.express as px
```

```python
# Q. How stocks performed over each sector for each year?
```

```python
# Reading the extracted csv files
df1 = pd.read_csv("Datasets/Auto_stock_data.csv")
df2 = pd.read_csv("Datasets/Financial_stock_data.csv")
df3 = pd.read_csv("Datasets/tech_data.csv")
df4 = pd.read_csv("Datasets/Healthcare_Stock_Data4.csv")
```

```python
# Concating all the dataframes
combined_df = pd.concat([df1, df2, df3,df4], ignore_index=True)
```

```python
#Converting the date column to datetime format in the final combined dataframe
combined_df['Date'] = pd.to_datetime(combined_df['Date'])
```

```python
# Creating a new column 'Year'
combined_df['Year'] = combined_df['Date'].dt.year
combined_df
```

```python
# Group the combined dataframe by Year and Sector on Average Volumne
grouped_df = combined_df.groupby(['Year', 'Sector'])['Volume'].mean().reset_index()

pivot_data = grouped_df.pivot(index='Year', columns='Sector', values='Volume').reset_index()
```

```python
# Plot the chart

plt.figure(figsize=(12, 8))
bar_width = 0.2  # Width of each bar

for i, sector in enumerate(grouped_df['Sector'].unique()):
    plt.bar(pivot_data['Year'] + i * bar_width, pivot_data[sector], width=bar_width, label=sector)

plt.xlabel('Year')
plt.ylabel('Average of Volume')
plt.title('Performance of Each Sector per year in terms of Average Volume')
plt.legend()
# plt.savefig("year_wise_sector_performance.png")
```

```python
# Read the csv file and save it in a dataframe
data_df1 = pd.read_csv("Datasets/Auto_stock_data.csv")
```

```python
# Convert date column to datetime
data_df1['Date'] = pd.to_datetime(data_df1['Date'])
```

```python
# Split the data industry wise
adf1 = data_df1.loc[data_df1['Industry']=="Auto & Truck Dealerships",:]
adf2 = data_df1.loc[data_df1['Industry']=="Auto Parts",:]
adf3 = data_df1.loc[data_df1['Industry']=="Auto Manufacturers",:]
adf4 = data_df1.loc[data_df1['Industry']=="Semiconductors",:]
```

```python
# Group each dataframe on Date column on Average Close price
agroup_df1 = adf1.groupby("Date")['Close'].mean().reset_index()
agroup_df2 = adf2.groupby("Date")['Close'].mean().reset_index()
agroup_df3 = adf3.groupby("Date")['Close'].mean().reset_index()
agroup_df4 = adf4.groupby("Date")['Close'].mean().reset_index()
```

```python
# Merge all the dataframes
amerged_df = pd.merge(agroup_df1, agroup_df2, on='Date', how='inner', suffixes=('_df1', '_df2'))
amerged_df = pd.merge(amerged_df, agroup_df3, on='Date', how='inner', suffixes=('_merged', '_df3'))
amerged_df = pd.merge(amerged_df, agroup_df4, on='Date', how='inner', suffixes=('_df3', '_df4'))
```

```python
# Rename all the columns to industry name
amerged_df = amerged_df.rename(columns={"Close_df1": "Auto & Truck Dealerships",
                                      "Close_df2": "Auto Parts",
                                      "Close_df3": "Auto Manufacturers",
                                      "Close_df4": "Semiconductors"})
```

```python
amerged_df = amerged_df.dropna()
```

```python
# Find the correlation between industries
close_columns1 = ['Auto & Truck Dealerships', 'Auto Parts', 'Auto Manufacturers', 'Semiconductors']
acorrelation_matrix = amerged_df[close_columns1].corr()
```

```python
# Plot a heatmap to show the correlation
plt.figure(figsize=(6,5))
sbn.heatmap(acorrelation_matrix, annot = True)
plt.xticks(rotation = 45, ha='right')
plt.title('Correlation of Closing Price')
# plt.savefig("automobile_correlation.png")
```

```python
# Find a linear regression
aslope, aintercept, arvalue, apvalue, astderr = st.linregress(amerged_df['Auto & Truck Dealerships'], amerged_df['Semiconductors'])

# Form the linear equation
aline_eq = f"y = {aslope:.2f}x + {aintercept:.2f}"

# Find the regression values based on the line equation
aregressed_values = (aslope * amerged_df['Auto & Truck Dealerships']) + aintercept

# Plot the scatter plot and linear regression
ax_axis = amerged_df['Auto & Truck Dealerships']
ay_axis = amerged_df['Semiconductors']
plt.scatter(ax_axis, ay_axis)
plt.plot(ax_axis, aregressed_values, color='r')
plt.annotate(aline_eq, (min(ax_axis), max(ay_axis)-100), fontsize=15, color='red')
plt.xlabel("Closing Price of Auto & Truck Dealerships")
plt.ylabel("Closing Price of Semiconductors")
acorrelation_coefficient = f'The r-value is: {arvalue:.2f}'
print(acorrelation_coefficient)
```

```python
fdata_df = pd.read_csv("Datasets/Financial_stock_data.csv")
```

```python
# Convert date column to datetime format
fdata_df['Date'] = pd.to_datetime(fdata_df['Date'])
```
