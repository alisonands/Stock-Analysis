```python
from alpha_vantage.timeseries import TimeSeries
import matplotlib.pyplot as plt
import sys
import pandas as pd

whole_data = []

symbols = ["ISRG","BDX","ALC","WST","RMD","UNH","ELV", "CVS","CI","HUM","NVO","VRTX","REGN","SGEN","MRNA","PFE","JNJ","LLY","AZN", "MRK"]

for symbol in symbols:
    ts = TimeSeries(key='9SJ9JBPTJZMWVIZW', output_format='pandas')
    data, meta_data = ts.get_daily(symbol=symbol, outputsize='full')
    data_df = pd.DataFrame(data)
    
# data_df.info()
    data_df = data_df.sort_index()
    
# Set the 'date' column as the index
    start_date = '2018-01-01'
    end_date = '2023-11-29'
    
    filtered_data = pd.DataFrame(data_df.loc[start_date:end_date])
# stock_data = data_date_changed

    filtered_data["Ticker"] = symbol
    whole_data.append(filtered_data)
    
print(whole_data)
```

```python
# Concatenate the list of dataframes into a single dataframe
result_df = pd.concat(whole_data)

# Save the dataframe to a CSV file
result_df.to_csv('Healthcare_Stock_Data.csv', index=True)
```
