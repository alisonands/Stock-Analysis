```python
from alpha_vantage.timeseries import TimeSeries
import matplotlib.pyplot as plt
import sys
import pandas as pd
import seaborn as sns

# Read CSV
healthcare_data = pd.read_csv("Datasets/Healthcare_Stock_Data4.csv")
automobiles_data = pd.read_csv("Datasets/final_auto_stock_data.csv")
finance_data = pd.read_csv("Datasets/Financial_stock_data.csv")
tech_data = pd.read_csv("Datasets/tech_data_update.csv")
```

```python
# Calculate the total volume for each industry
total_volume_by_industry = healthcare_data.groupby('Industry')['Volume'].sum()

# Calculate the percentage of total volume for each industry
volume_percentage_by_industry = (total_volume_by_industry / total_volume_by_industry.sum()) * 100

# Define colors for each industry
colors = plt.cm.Paired(range(len(volume_percentage_by_industry)))

# Explode a section for emphasis 
explode = [0.1 if i == volume_percentage_by_industry.idxmax() else 0 for i in volume_percentage_by_industry.index]

# Plot a pie chart
plt.figure(figsize=(10, 6))
plt.pie(volume_percentage_by_industry, labels=volume_percentage_by_industry.index, autopct='%1.1f%%', startangle=140,
        colors=colors, explode=explode, shadow=True)
plt.title('Portfolio Allocation by Industry based on Volume (Sector - Healthcare)')

# Save the pie chart as a PNG file 
plt.savefig('Healthcare_pie_chart.png')


plt.show()
```

```python


# Merge all datasets on 'Sector' column
merged_data = pd.concat([healthcare_data, automobiles_data, finance_data, tech_data])

# Set the style
sns.set(style="whitegrid")

# Calculate the total volume for each sector
total_volume_by_sector = merged_data.groupby('Sector')['Volume'].sum()

# Calculate the percentage of total volume for each sector
volume_percentage_by_sector = (total_volume_by_sector / total_volume_by_sector.sum()) * 100

# Define colors for each sector
colors = sns.color_palette("pastel", len(volume_percentage_by_sector))

# Explode a section for emphasis 
explode = [0.1 if i == volume_percentage_by_sector.idxmax() else 0 for i in volume_percentage_by_sector.index]

# Plot a pie chart with enhanced styling
plt.figure(figsize=(12, 8))
plt.pie(volume_percentage_by_sector, labels=volume_percentage_by_sector.index, autopct='%1.1f%%', startangle=140,
        colors=colors, explode=explode, shadow=True, wedgeprops=dict(width=0.4, edgecolor='w'))

# Add a legend
plt.legend(title="Sector", loc="upper left", bbox_to_anchor=(1, 0, 0.5, 1))

# Add a title
plt.title('Grand Portfolio Allocation by Sector based on Volume', fontsize=16)

# Remove y-axis label for cleaner look
plt.ylabel('')

# Save the pie chart as a PNG file 
plt.savefig('grand_pie_chart.png')


# Display the plot
plt.show()
```
